## Example 1.10. 
Let us plot the decimal representation of the numbers in the binary system spec- ified by $(\beta,t,L,U) = (2,2,−2,3)$. 

The smallest possible number in the significand $1.d_1d_2$ is $1.00$ and the largest possible number is $1.11$ in binary, which is equal to $1 + 1/2 + 1/4 = 1.75$. 

So, we will run in a loop from $1$ to $1.75$ in increments of $\beta−t = 2−2 = 0.25$. 

The exponent runs from $−2$ to $3$, which yields a second, nested loop. The resulting double loop builds up an array with all the positive elements of the system.

Here is a Julia script that plots the numbers of the system. Notice the format in the plot command.

```julia
using Plots

x = [] #empty array

# Generate all positive numbers of the system (2, 2, −2, 3)
for i in 1:0.25:1.75
    for j in -2:3
        push!(x, i * (2.0)^j) # "creating" new element 
    end
end

# Add all negative numbers and 0
x = vcat(x, -x, 0)  #vertical concatenation

# Sort the array
x = sort(x)

# Create y values (zeros) for each x value
y = zeros(length(x))

# Plot with '+' markers and set y-axis limits
plot(x, y, seriestype = :scatter, marker = :+, legend = false, ylims = (-1, 1))
```
