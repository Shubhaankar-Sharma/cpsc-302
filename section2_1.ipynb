## Example 2.2

Here is the Julia script that generates Figure 2.1 given in the textbook.


```julia
using Plots, Roots

# Generate plot for f_1
x = 0:0.1:4*pi
fx = sin.(x)
zx = zeros(length(x))
xrt = 0:pi:4*pi
yrt = zeros(5)
plot(x, fx, label="f_1(x)", color=:blue)
plot!(x, zx, label="", color=:black)
scatter!(xrt, yrt, color=:red, marker=:x,label=:"roots")
xlims!(0, 4*pi)
ylims!(-1, 1)
xlabel!("x")
ylabel!("f_1(x)")

```

```julia
# Generate plot for f_2
x = 0:0.1:20
fx = x.^3 .- 30*x.^2 .+ 2552
zx = zeros(length(x))
xrt = find_zero(x -> x^3 - 30*x^2 + 2552, 10)
plot(x, fx, label="f_2(x)", color=:green)
plot!(x, zx, label="", color=:black)
scatter!([xrt], [0], color=:red, marker=:x,label=:"roots")
xlims!(0, 20)
ylims!(-2000, 3000)
xlabel!("x")
ylabel!("f_2(x)")
```

```julia
# Generate plot for f_3
x = -10:0.1:10
fx = 10*cosh.(x/4) .- x
zx = zeros(length(x))
plot(x, fx, label="f_3(x)", color=:red)
xlims!(-10, 10)
ylims!(0, 100)
xlabel!("x")
ylabel!("f_3(x)")
```
