```julia
e = exp(1);
n = 1:10;
Sn = sqrt.(2*pi*n).*((n/e).^n);
# Sn = sqrt(2*pi*n).*((n/e).^n);
fact_n = factorial.(n);
abs_err = abs.(Sn-fact_n);
rel_err = abs_err ./ fact_n;
# format short g
[n; fact_n; Sn; abs_err; rel_err]'
```

```julia
z = LinRange(0,1,101);
exacterf = erf.(z);
approxF(z) = (2/sqrt(pi))*(z-(1/3)*(z^3));
approxerf = approxF.(z);
```

```julia
using Plots

plot(z, exacterf, label="Exact");
plot!(z, approxerf, label="Approx");
xlims!(0, 1)  # Set x-axis limits to 0 and 1
```

```julia
abs_err = abs.(exacterf-approxerf);
semilogy(z, abs_err);
```
